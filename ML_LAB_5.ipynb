```python
import math
ep_cnt=0

def step(x):
    y = math.modf(x)
    return y[1]


def ep(w0, w1, w2, x1, x2, t, n, ep_cnt):
    f = False

    for i in range(4):
        yin = float(w1 * x1[i] + x2[i] * w2 + w0)

        if step(yin) == t[i]:
            continue
        else:
            print("updating ")
            w1 = w1 + n * (t[i] - step(yin)) * x1[i]
            w2 = w2 + n * (t[i] - step(yin)) * x2[i]
            w0 = w0 + n * (t[i] - step(yin))
            f = True

        print(f"Weights are in  {i + 1} iteration ")
        print(f"W1 = {w1}")
        print(f"W2 = {w2}")
        print(f"W0 = {w0}")

    if f:
        print("updating weights are of previous epochs")
        print(f"epoch {ep_cnt}")
        ep_cnt+=1
        ep(w0, w1, w2, x1, x2, t, n, ep_cnt + 1)
    else:
        print("Weights are not changed ")
        print("Final weights are: ")
        print(f"W1 = {w1}")
        print(f"W2 = {w2}")
        print(f"W0 = {w0}")


w0 = 10
w1 = 0.2
w2 = -0.75
n = 0.05
x1 = [0, 0, 1, 1]
x2 = [0, 1, 0, 1]
t = [0, 0, 0, 1]

ep(w0, w1, w2, x1, x2, t, n, 1)
```
